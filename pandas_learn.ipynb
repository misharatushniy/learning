```python
%matplotlib inline
import matplotlib as plt
import seaborn as sns

student_performans = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
student_performans
```

```python
student_performans = student_performans.rename(columns={'math score':'math_score', 'reading score':'reading_score'})
student_performans.math_score.hist()
```

```python
student_performans.plot.scatter(x='math_score', y='reading_score')
```

```python
sns.set(style="darkgrid")
ax = sns.lmplot(x='math_score', y='reading_score', hue='gender', data=student_performans, fit_reg=False)
ax.set_xlabels('Math score')
ax.set_ylabels('Reading score')
```
