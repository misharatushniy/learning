```python
m_dict = {
        "января":0,
        "февраля":31,
        "марта":59,
        "апреля":90,
        "мая":120,
        "июня":151,
        "июля":181,
        "августа":212,
        "сентября":243,
        "октября":273,
        "ноября":304,
        "декабря":334
         }
```

```python
import csv

# чтение с тхт и запись в csv удаляя все лишнее
with open('data_with_int_mounths.csv', 'w', encoding='utf-8') as data:
    file_writer = csv.writer(data, delimiter = ",", lineterminator="\r")
    with open('soup.txt', 'r') as soup_txt:
        for line in soup_txt:
            a = line.strip().split()
            del a[6:8]   # Удаляем строку "г."
            del a[-1]     
            try:     # переводим все числовые значения в инт
                if a[3].isdigit():
                    a[3] = int(a[3])
                    a[5] = int(a[5])
                    a[6] = int(a[6])
                    a[8] = int(a[8])
                    a[4] = m_dict[a[4]]
                    a[7] = m_dict[a[7]]
                    file_writer.writerow(a) # запись каждой строчки из тхт в  csv
                else:
                    None
            except:
                None
```
