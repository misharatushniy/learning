```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://pomnim.online/cemetery/2099/page/'
page = 1
r = requests.get(url)
soup = BeautifulSoup(r.text, features='lxml')
```

```python
people = soup.find_all('div', class_='burial-teaser__content flex-grow-1')
```

```python
for i in people:
    print(i.find('a', class_="burial-teaser__title").text, i.find('span').text)
```

```python
import time
page = 100
data = []
while page != 650:
    url = 'https://pomnim.online/cemetery/2099/page/' + str(page)
    r = requests.get(url)
    time.sleep(15)
    soup = BeautifulSoup(r.text, features='lxml')
    people = soup.find_all('div', class_='burial-teaser__content flex-grow-1')
    for i in people:
        data.append(i.find('a', class_="burial-teaser__title").text + ' ' + i.find('span').text)
    page += 1
    print(url)
    print(len(data))
with open('soup.txt', 'a') as file:
    for i in data:
        file.write(i + '\n')
```
